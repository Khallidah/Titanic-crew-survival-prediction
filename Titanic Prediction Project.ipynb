```python
### Dropping the columns we do not need and then assigning it to a new variable name new_df
```

```python
new_df = df.drop(['PassengerId','Name','Ticket'],axis =1)
new_df
```

## CHECKING MISSING VALUES 

```python
new_df.isnull().sum()
```

### CONVERTING CATEGORICAL FEATURES TO NUMERICAL FEATURES

```python
new_df['Sex'].unique()
```

```python
### Dealing with the sex column and using the map function/method

new_df['Sex'] = new_df['Sex'].map({'male':0,'female':1})
```

```python
new_df
```

```python
### Dealing with the cabin column

from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le = LabelEncoder()
new_df['Cabin'] = le.fit_transform(new_df['Cabin'])
new_df.info()
```

```python
new_df
```

### When we used the labelEncoder on the cabin column it automatically converted all the categorical variables in that column including the nan values to numerical features

```python
### lets deal with the Embarked column using the get dummies variable
emb_cat = new_df['Embarked'].value_counts()
emb_cat.index
```

```python
most_occuring_embarked = emb_cat.index[0]
new_df['Embarked'] = new_df['Embarked'].fillna(most_occuring_embarked)

emb = pd.get_dummies(new_df['Embarked'],drop_first = True, prefix = 'Embarked')
emb
```

##### Now we have used our getdummy variable but then dont you forget we assigned it to a variable name called emb but our original dataset is new_df so we'd have to concatenate them together

```python
new_df = pd.concat([new_df,emb],axis = 1 )

### We have to drop the Embarked since we do not need it anylonger 

new_df.drop('Embarked',axis = 1, inplace = True)
new_df
```

#### Dealing with our age column now lets fill it up with the mean..

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
```

```python
new_df.isnull().sum()
```

```python
new_df.info()
```

### Now at this point we have dealt with our categorical features by converting them to  numerical features also we dealt missing values so we can now proceed to building our model

```python
y = new_df['Survived']
X = new_df.drop('Survived',axis =1)
```

```python
### train_set_split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
### Fitting my model to the train file
logreg.fit(X_train,y_train)
```

```python
pred = logreg.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred) * 100
```

```python
## pickle
import pickle

pickle_out = open('model.pkl', 'wb')
pickle.dump(logreg, pickle_out)
pickle_out.close()
```
